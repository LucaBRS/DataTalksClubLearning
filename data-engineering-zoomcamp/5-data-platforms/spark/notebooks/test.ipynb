```python
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('intro-to-pyspark').getOrCreate()
df = spark.read.parquet('fhv_tripdata_2021-02.parquet')
df.limit(100).write.parquet('head.parquet', mode='overwrite')

df.show()
```

```python
import pandas as pd
df_pandas = pd.read_parquet('head.parquet')
df_pandas.dtypes
```
