```python
# SET UP TABLE FOR FLINK
import psycopg2
conn = psycopg2.connect(
    host="localhost",
    database="postgres",
    user="postgres",
    password="postgres"
)

cur = conn.cursor()

cur.execute("""
CREATE TABLE IF NOT EXISTS trip_counts (
    window_start TIMESTAMP,
    PULocationID INTEGER,
    num_trips BIGINT
    )
    """)

conn.commit()
cur.close()
conn.close()
print("Table created successfully.")
```

```python
import psycopg2

conn = psycopg2.connect(
    host="localhost",
    port=5432,
    database="postgres",
    user="postgres",
    password="postgres"
)
cur = conn.cursor()

cur.execute("""
    SELECT PULocationID, num_trips
    FROM trip_counts
    ORDER BY num_trips DESC
    LIMIT 3
""")

for row in cur.fetchall():
    print(row)

cur.close()
conn.close()
```
