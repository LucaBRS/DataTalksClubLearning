```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName('intro') \
    .getOrCreate()

spark
```

## Creare un DataFrame da dati inline

```python
data = [
    (1, 'Alice', 30),
    (2, 'Bob',   25),
    (3, 'Carol', 35),
]

df = spark.createDataFrame(data, schema=['id', 'name', 'age'])
df.show()
```

## Schema e tipi

```python
df.printSchema()
```

## Filtri e selezione

```python
df.filter(df.age > 28).select('name', 'age').show()
```

## Aggregazioni

```python
from pyspark.sql import functions as F

df.agg(
    F.avg('age').alias('avg_age'),
    F.max('age').alias('max_age'),
).show()
```

## SQL su DataFrame

```python
df.createOrReplaceTempView('persone')

spark.sql('SELECT name, age FROM persone WHERE age > 28').show()
```
