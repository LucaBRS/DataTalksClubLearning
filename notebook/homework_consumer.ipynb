```python
# CREATING KAFKA CONSUMER
import json
import sys
sys.path.append("../src")

from kafka import KafkaConsumer
from models import Ride, ride_deserializer
import psycopg2

# cpostgress connection (generic way to connect to postgres, not specific to kafka)
conn = psycopg2.connect(
    host="localhost",
    port=5432,
    database="postgres",
    user="postgres",
    password="postgres"
)
cur = conn.cursor()


def json_deserializer(data):
    return json.loads(data.decode('utf-8'))


# cpConnection to kafka consumer
consumer = KafkaConsumer(
    "green-trips",
    bootstrap_servers="localhost:19092",
    value_deserializer=json_deserializer,
    auto_offset_reset="earliest",
    group_id="homework"
)

print("Consumer pronto!")
```

```python
import pandas as pd
dfs = []
calc_flag = True
while True:
    msg = consumer.poll(timeout_ms=1000)
    if msg:
        calc_flag = True
        for tp, messages in msg.items():
            for message in messages:
                ride_dict = message.value
                dfs.append(pd.DataFrame([ride_dict]))
    else:

        if dfs and calc_flag:
            calc_flag = False
            df = pd.concat(dfs, ignore_index=True)
            df.drop_duplicates(inplace=True)
            print(df[df['trip_distance']>5].shape[0])
```
